```python
# 
class Solution:
    def letterCombinations(self, digits: str) -> list:
        if not digits:
            return []
        num2strs = {'2': 'abc', '3': 'def', '4': 'ghi', '5': 'jkl', '6': 'mno', '7': 'pqrs', '8': 'tuv', '9': 'wxyz'}
        results = []
        def backtracking(res, inx):
            if inx==len(digits):
                results.append(res)
                return
            num = digits[inx]
            strs = num2strs[num]
            for c in strs:
                res += c
                backtracking(res, inx+1)
                res = res[:-1]
                # backtracking(res+c, inx+1)
        backtracking('', 0)
        return results

s = Solution()
s.letterCombinations('23')
```
