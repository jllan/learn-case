## 电话号码的字母组合

给定一个仅包含数字 2-9 的字符串，返回所有它能表示的字母组合。
给出数字到字母的映射如下（与电话按键相同）。注意 1 不对应任何字母。


示例:
>输入："23"  
输出：["ad", "ae", "af", "bd", "be", "bf", "cd", "ce", "cf"].

```python
# 解法一：循环
class Solution:
    def letterCombinations(self, digits: str) -> list:
        if not digits:
            return []
        num2strs = {'2': 'abc', '3': 'def', '4': 'ghi', '5': 'jkl', '6': 'mno', '7': 'pqrs', '8': 'tuv', '9': 'wxyz'}
        state = ['']
        for d in digits:
            strs = num2strs[d]
            state_new = []
            for c in strs:
                for s in state:
                    state_new.append(s+c)
            state = state_new
        return state
```

```python
s = Solution()
s.letterCombinations('23')
```

```python
# 解法二：回溯
class Solution:
    def letterCombinations(self, digits: str) -> list:
        if not digits:
            return []
        num2strs = {'2': 'abc', '3': 'def', '4': 'ghi', '5': 'jkl', '6': 'mno', '7': 'pqrs', '8': 'tuv', '9': 'wxyz'}
        results = []
        def backtracking(res, inx):
            if inx==len(digits):
                results.append(res)
                return
            num = digits[inx]
            strs = num2strs[num]
            for c in strs:
                res += c
                backtracking(res, inx+1)
                res = res[:-1]
                # backtracking(res+c, inx+1)
        backtracking('', 0)
        return results

s = Solution()
s.letterCombinations('23')
```
